```python
#import pandas
import pandas as pd

#Import Scraping Library
from google_play_scraper import app, reviews
```

## Scraping Data

```python
# Menentukan ID aplikasi
app_id = 'co.id.bankbsi.superapp'

# Mengambil ulasan
reviews_data, _ = reviews(
    app_id,
    lang='id',
    country='id',
    count=10000
)

# Konversi ke DataFrame
reviews_df = pd.DataFrame(reviews_data)
reviews_df
```

```python
# save to csv
reviews_df.to_csv('bsi_reviews.csv', index=False)
```
